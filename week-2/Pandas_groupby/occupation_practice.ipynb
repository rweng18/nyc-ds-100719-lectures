### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep = '|')
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation')['age'].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
# Group dataframe by Occupation and Gender, find counts, reset index so columns are callable
gender = users.groupby(['occupation','gender'])['user_id'].count().reset_index()

# Find ratio by gender counts, grouped by occupation
gender['ratio'] = gender.groupby(['occupation'])['user_id'].apply(lambda x: x/x.sum())

# Take only male ratios
gender_males = gender[gender['gender'] == 'M']

# Sort
gender_males_sorted = gender_males.sort_values('ratio', ascending=False)
gender_males_sorted
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
ages = users.groupby(['occupation'])['age'].agg(['max', 'min'])
ages
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
ages = users.groupby(['occupation', 'gender'])['age'].agg('mean')
ages
```

### Step 8.  For each occupation present the percentage of women and men

```python
gender = users.groupby(['occupation','gender'])['user_id'].count().reset_index()
gender
gender['percent'] = gender.groupby(['occupation'])['user_id'].apply(lambda x: round(x/x.sum() * 100, 2))
gender
```
